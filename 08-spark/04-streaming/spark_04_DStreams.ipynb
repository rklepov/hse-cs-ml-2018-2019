### Spark Streaming: DStreams (_deprecated_)

```python
import findspark
findspark.init()

import pyspark
```

```python
sc = pyspark.SparkContext()
```

```python
from pyspark.streaming import StreamingContext
```

```python
ssc = StreamingContext(sparkContext=sc, batchDuration=10)
```

```python
ssc.checkpoint(directory="/tmp")
```

```python
stream_in = "/stream/legacy"

!mkdir -p $stream_in
```

```python
input_stream = ssc.textFileStream(directory=stream_in)
```

```python
output_stream = (
    input_stream
    .map(int)
    .map(lambda x: 'odd' if x % 2 else 'even')
    .countByValue()
)
```

```python
transfor_stream = (
    input_stream
    .map(int)
    .map(lambda x: 'odd' if x % 2 else 'even')
    .map(lambda x: (x, 1))
)
```

```python
output_stream.pprint()
```

```python
transfor_stream.groupByKeyAndWindow(windowDuration=50, slideDuration=20).mapValues(sum).pprint()
```

```python
ssc.start()
```
