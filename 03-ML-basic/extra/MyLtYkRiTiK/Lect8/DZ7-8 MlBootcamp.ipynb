## Задание. Срок: 14 декабря.

В этом задании вам нужно будет исследовать данные по сердечно-сосудистым заболеваниям. Подробнее о данных и задаче можно посмотреть на страничке соревнования: [mlbootcamp.ru](https://mlbootcamp.ru/round/12/sandbox/). Данные можно скачать по ссылке, а можно найти в гите в папке 'data/ml5/'.

Соревнование проходило несколько лет назад, но сейчас есть возможность отправлять решения и смотреть скор в режиме "песочницы".

Итак, что нужно сделать:
1. Это соревнование отличалось большим количеством сорных или аномальных данных. Вам нужно постараться что-то с ними сделать - локализовать и исправить, или удалить, или оставить как есть.
2. Построить, допустим, быстрый Ridge на данных с аномалиями и с исправленными аномалиями.
3. Построить ANN на обычных данных и данных с аномалиями. Подобрать гиперпараметры. Сравнить результаты с работой риджа. Для проверки скора я бы советовала валидацию, но вы можете выбирать любой сбособ, главное обосновать, какая модель лучше и почему.
4. (Опционально) Заслать свои решения в песочницу конкурса и посмотреть, а что же лидерборд думает про ваше решение.
