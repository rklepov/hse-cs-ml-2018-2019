# Matplotlib Object Oriented Method
Давайте познакомимся с более формальным введением в Matplotlib's Object Oriented API. Это значит, что мы будем создавать figure объекты и затем вызывать методы или атрибуты этих объектов.

```python
# Создаем пустую канву
fig = plt.figure()

# Добвляем оси
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
# x0, y0, ширина, высота (от 0 до 1) - однозначно определяет положение и размер создаваемого рисунка

axes.plot(x, y, 'b')
axes.set_xlabel('Задать X ярлыки') # Мы используем set_, чтобы начать метод
axes.set_ylabel('Задать y ярлыки')
axes.set_title('Задать заголовок')
plt.show()
```

Контроль над положениями значит, что мы можем делать такие вещи:

```python
# Создаем пустую канву
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # основные оси
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # добавленные оси

axes1.plot(x, y, 'b')
axes1.set_xlabel('X_ярлыки_axes2')
axes1.set_ylabel('Y_ярлыки_axes2')
axes1.set_title('Заголовок 1')

axes2.plot(y, x, 'r')
axes2.set_xlabel('X_ярлыки_axes2')
axes2.set_ylabel('Y_ярлыки_axes2')
axes2.set_title('Заголовок 2');
```

## subplots()

plt.subplots() представляет из себя более автоматизированный менеджер осей.

```python
# Аналогично plt.figure(), только возвращает пару канвы и осей
fig, axes = plt.subplots()

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Заголовок');
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
for ax in axes:
    ax.plot(x, y, 'b')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('Заголовок')

fig
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
    ax.plot(x, y, 'g')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('Заголовок')

fig    
plt.tight_layout()
```

### Размер, соотношение сторон и DPI

Размер, соотношение сторон и DPI могут быть заданы при создании канвы. 
* `figsize` - это пара ширины и высоты канвы в дюймах
* `dpi` - это dots-per-inch. 

```python
fig = plt.figure(figsize=(8,4), dpi=100)
```

Эти же аргументы могут быть переданы в функцию `subplots`:

```python
fig, axes = plt.subplots(figsize=(12,3))

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Заголовок')
```

```python
fig, axes = plt.subplots(figsize=(12,3))

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Заголовок')
axes.axvline(x=4) # добавим вертикальную линию
```

## Цвета, типы линий и их ширина

```python
fig, ax = plt.subplots()
ax.plot(x, x**2, 'b.-') # голубая линия с точками
ax.plot(x, x**3, 'g--') # зеленая пунктирная линия
```

```python
fig, ax = plt.subplots()

ax.plot(x, x+1, color="blue", alpha=0.5) # полупрозрачная линия
ax.plot(x, x+2, color="#8B008B")        # RGB hex code
ax.plot(x, x+3, color="#FF8C00")        # RGB hex code 
```

Для изменения ширины линии можно использовать `linewidth` или `lw` аргументы. Для стиля строки `linestyle` `ls`.

```python
fig, ax = plt.subplots()

ax.plot(x, x+1, color="blue", alpha=0.5, lw=3, linestyle='--') # полупрозрачная линия
ax.plot(x, x+2, color="#8B008B", linewidth=0.25)        # RGB hex code
ax.plot(x, x+3, color="#FF8C00", lw=3, ls='-', marker='s')        # RGB hex code 
```

### Пример для самостоятельного разбора

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)

# возможные опции типа линии: ‘-‘, ‘–’, ‘-.’, ‘:’, ‘steps’
ax.plot(x, x+5, color="green", lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')

# кастомное пунктирование
line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # формат: line length, space length, ...

# возможные опции маркеров: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')

# размер маркера и цвет
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```
