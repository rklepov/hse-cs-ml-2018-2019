#### Машинный ноль:
Максимальное положительное eps, для которого справедливо равенство 1+eps=1. (Википедия)

```python
eps = 1
while 1.0 + eps != 1.0:
    print('diff is smaller than', eps)
    eps = eps/2
print('finally!', eps)
```
